# Section 2 Data Cleaning

Section 2 content i will be checking the data for any duplocates, missing values, and any inconsistencies. The data will be getting prepared to be saved as a clean dataset ready for visualisation. 

isnull().sum is used to count the number of missing values in each column of a dataset. It helps identify which columns contain null or empty entries and shows if data cleaning is needed. It was used to check the dataset for missing values. The dataset does not have any missing values so will need to fill in any missing values.However, If missing values had been found, they would have been handled by filling them with appropriate values.

```python
# Checking to see if there are any missing values in the dataset
df.isnull().sum()
# There are no missing values in the dataset, which is good because it means that all the data is complete and there are no gaps in the information.
```

.duplicated() is used to find duplicate rows in the dataset.This dataset does not have any duplicates but if there were duplicates, they would have been removed by using the .drop_duplicates function.

```python
#Checking for duplicate rows in the dataset
df.duplicated().sum()
# this shows that there are no duplicate rows in the dataset, which is good because it means that each row represents a unique customer.
```

# Dropping columns
The .drop function drops any columns which are specified. it used to drop any columns which will not be useful in further analysis. In this case dropping the 'CLIENTNUM' colum as it is not of any use because it is just an ID coulmn. axis=1 is written to drop the column and inplace=True to overwrite from the original dataset.
df.head() is used to check if the column has been dropped.

```python
# i am going to drop the 'CLIENTNUM' column from the dataset because it is not relevant to the analysis and will not be used in the machine learning models. The 'CLIENTNUM' column is a unique identifier for each customer and is not going to be useful in my further analysis
df.drop('CLIENTNUM', axis=1, inplace=True)
df.head() # i will show the first 5 rows of the dataset to see the changes made to the dataset. The 'CLIENTNUM' column has been dropped from the dataset.
```

2 further columns have been dropped as they wont be needed for data visualisation and cleans up the dataset

```python
# Dropping 2 furth columns which are of no use in the analysis.
df.drop(['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], axis=1, inplace=True)
df.head() # Checks the first 5 rows of the dataset to see the changes.
```

# Encoding categorical values
By using the .map function is has allows you to convert text values into numbers for selected columns. It has been created beauce machine learning models cannot work with text data as it will crash the code so it has been converted into numbers. In this case, we have used the 'Attrition_Flag' column and coverted 'Existing Customer' to 0 and 'Attrited Customer' to 1. 
df.head is used to check the code has worked
the 'Attrition_Flag' column is now ready to be used in machine learning. 

```python
#i am going to encode the categorical variables in the dataset using one-hot encoding. This will convert the categorical variables into numerical variables that can be used in machine learning models. i will be encoding the 'attrition_flag'
df['Attrition_Flag'] = df['Attrition_Flag'].map({'Existing Customer': 0, 'Attrited Customer': 1})
df.head() # i will show the first 5 rows of the dataset to see the changes made to the 'Attrition_Flag' column. The values in the 'Attrition_Flag' column have been converted from categorical values to numerical values, with 'Existing Customer' being represented as 0 and 'Attrited Customer' being represented as 1.
```

# Saving the cleaned dataset in the cleaned data folder

```python
df.to_csv('Dataset/CleanedData/BankChurners_cleaned.csv', index=False) # i will save the cleaned dataset to a new csv file called 'BankChurners_cleaned.csv' in the 'CleanedData' folder. The index=False parameter is used to prevent the index column from being saved to the csv file.
print('Cleaned dataset has been saved to CleanedData folder!')
```
