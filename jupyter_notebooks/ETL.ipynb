# Section 2 Data Cleaning

Section 2 content i will be checking the data for any duplocates, missing values, and any inconsistencies, i will adjust accordingly

```python
# checking to see if there are any missing values in the dataset
df.isnull().sum()
# there are no missing values in the dataset, which is good because it means that all the data is complete and there are no gaps in the information.
```

```python
#Checking for duplicate rows in the dataset
df.duplicated().sum()
# this shows that there are no duplicate rows in the dataset, which is good because it means that each row represents a unique customer.
```

```python
#i am going to encode the categorical variables in the dataset using one-hot encoding. This will convert the categorical variables into numerical variables that can be used in machine learning models. i will be encoding the 'attrition_flag'
df['Attrition_Flag'] = df['Attrition_Flag'].map({'Existing Customer': 0, 'Attrited Customer': 1})
df.head() # i will show the first 5 rows of the dataset to see the changes made to the 'Attrition_Flag' column. The values in the 'Attrition_Flag' column have been converted from categorical values to numerical values, with 'Existing Customer' being represented as 0 and 'Attrited Customer' being represented as 1.
```

```python
# i am going to drop the 'CLIENTNUM' column from the dataset because it is not relevant to the analysis and will not be used in the machine learning models. The 'CLIENTNUM' column is a unique identifier for each customer and is not going to be useful in my further analysis
df.drop('CLIENTNUM', axis=1, inplace=True)
df.head() # i will show the first 5 rows of the dataset to see the changes made to the dataset. The 'CLIENTNUM' column has been dropped from the dataset.
```
