# Section 1

Section 1 content i will be importing my dataset in

```python
# loading in the raw CSV file
df = pd.read_csv('Dataset/RawData/BankChurners.csv')
df.head()   #showing the first 5 rows of the dataset
```

```python
#checking the shape of the dataset it will show the number of rows and columns in the dataset
df.shape
#it shows that the dataset has 10127 rows and 23 columns
```

```python
# Checking the names of the columns in the dataset
df.columns
```

```python
# Checking the data types of each column in the dataset
df.dtypes
```

```python
df.info()  #this will show the data types of each column and if there are any null values in the dataset
```

```python
# using the describe() function to get a statistical summary of the 'months on book' and 'credit limit' columns in the dataset
df[['Months_on_book', 'Credit_Limit']].describe().round(2)
#i have told the function to round the values to 2 decimal places for better readability. The output will show the count, mean, standard deviation, minimum, maximum, and quartile values for both columns.
```

```python
#
df['Attrition_Flag'].unique()  #this will show the unique values in the 'Attrition_Flag' column
#this shows that the 'Attrition_Flag' column has two unique values: 'Existing Customer' and 'Attrited Customer'
```

```python
# now i will be checking the value counts of the 'Attrition_Flag' column to see how many customers are existing and how many have attrited
df['Attrition_Flag'].value_counts()
# this shows that there are 8500 existing customers and 1627 attrited customers in the dataset. This means that the majority of the customers in the dataset are existing customers, while a smaller portion have attrited.
```
